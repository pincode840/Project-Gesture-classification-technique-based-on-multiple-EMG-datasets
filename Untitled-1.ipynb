```python
import pandas as pd
import numpy as np
# import torch  <-- 에러의 원인이므로 주석 처리

def create_dataset_numpy_only(df, window_size=500, overlap_ratio=0.5, ignore_rest=True):
    """
    PyTorch 없이 Numpy만 사용하여 대용량 데이터를 전처리하는 함수
    """
    print("1. 데이터프레임을 Numpy로 변환 중...")
    # 메모리 절약을 위해 float32 사용
    # iloc[:, :16]은 emg1 ~ emg16 컬럼을 가정
    data_np = df.iloc[:, :16].values.astype(np.float32) 
    labels_np = df['Restimulus'].values.astype(np.int32)
    
    print("2. 동작 구간(Block) 분할 중...")
    # 라벨이 변하는 구간 탐지
    change_indices = np.where(np.diff(labels_np, prepend=labels_np[0]-1) != 0)[0]
    change_indices = np.append(change_indices, len(labels_np))
    
    X_list = []
    Y_list = []
    
    step_size = int(window_size * (1 - overlap_ratio))
    
    print(f"3. 슬라이딩 윈도우 수행 중 (Window: {window_size}, Step: {step_size})...")
    
    for i in range(len(change_indices) - 1):
        start_idx = change_indices[i]
        end_idx = change_indices[i+1]
        
        current_label = labels_np[start_idx]
        
        # 휴식(0) 제외
        if ignore_rest and current_label == 0:
            continue
            
        block_data = data_np[start_idx:end_idx]
        
        if len(block_data) < window_size:
            continue
            
        # 블록 내부 윈도우 슬라이싱
        for w_start in range(0, len(block_data) - window_size + 1, step_size):
            window = block_data[w_start : w_start + window_size]
            
            # (Time, Channel) -> (Channel, Time) 전치
            X_list.append(window.T) 
            Y_list.append(current_label)
            
    print("4. 결과 병합 중...")
    X_final = np.array(X_list, dtype=np.float32)
    Y_final = np.array(Y_list, dtype=np.int32)
    
    # 라벨 0부터 시작하도록 조정 (만약 1부터 시작한다면)
    if len(Y_final) > 0 and Y_final.min() > 0:
        Y_final = Y_final - 1
        
    print(f"완료! X shape: {X_final.shape}, Y shape: {Y_final.shape}")
    return X_final, Y_final

# ==========================================
# 실행 부분
# ==========================================

# 1. 함수 실행 (Torch 임포트 없이 진행)
x_train_np, y_train_np = create_dataset_numpy_only(
    learning_df, 
    window_size=500, 
    overlap_ratio=0.5
)

# 2. 결과 저장 (중요: PyTorch 고칠 때까지 데이터 보존)
np.save('x_train.npy', x_train_np)
np.save('y_train.npy', y_train_np)
print("데이터가 .npy 파일로 저장되었습니다.")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
import time

# ==========================================
# 1. 모델 정의 (CPU용 경량화 버전)
# ==========================================
class ChannelAgnosticNetwork(nn.Module):
    def __init__(self, num_classes, d_model=64, input_channels=16): # d_model을 128->64로 줄임 (속도 향상)
        super(ChannelAgnosticNetwork, self).__init__()
        
        # [Stage 1] Shared Encoder 
        self.shared_encoder = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=11, padding=5), # 필터 수 줄임
            nn.BatchNorm1d(16), nn.ReLU(), nn.MaxPool1d(2),
            
            nn.Conv1d(16, 32, kernel_size=5, padding=2),
            nn.BatchNorm1d(32), nn.ReLU(), nn.MaxPool1d(2),
            
            nn.Conv1d(32, d_model, kernel_size=3, padding=1),
            nn.AdaptiveAvgPool1d(1), 
            nn.Flatten()             
        )
        
        # [Stage 2] Attention Mechanism
        self.attn_fc = nn.Sequential(
            nn.Linear(d_model, d_model // 2),
            nn.Tanh(),
            nn.Linear(d_model // 2, 1)
        )
        
        # Classifier
        self.classifier = nn.Sequential(
            nn.Linear(d_model, 32),
            nn.ReLU(),
            nn.Dropout(0.3), # Dropout 낮춤
            nn.Linear(32, num_classes)
        )

    def forward(self, x, mask=None):
        B, C, T = x.shape
        x_reshaped = x.view(B * C, 1, T)
        
        features = self.shared_encoder(x_reshaped) 
        features = features.view(B, C, -1)         
        
        attn_scores = self.attn_fc(features).squeeze(-1) 
        
        if mask is not None:
            attn_scores = attn_scores.masked_fill(mask == 0, -1e9)
            
        attn_weights = torch.softmax(attn_scores, dim=1).unsqueeze(-1) 
        global_feature = torch.sum(features * attn_weights, dim=1) 
        
        return self.classifier(global_feature)

# ==========================================
# 2. 학습 실행 함수 (CPU 최적화)
# ==========================================
def train_model_cpu():
    # --- CPU 강제 설정 ---
    DEVICE = torch.device("cpu")
    print(f"✅ 현재 설정된 장치: {DEVICE} (그래픽카드 없이 학습합니다)")
    
    # 설정값 (CPU 부하를 줄이기 위해 배치 사이즈 조절)
    BATCH_SIZE = 32 # 64 -> 32 (메모리 부담 완화)
    LEARNING_RATE = 0.001
    EPOCHS = 10     # 테스트를 위해 10으로 설정

    # --- 데이터 로드 ---
    print("📂 데이터 로드 중...")
    try:
        X_all = np.load('x_train.npy')
        Y_all = np.load('y_train.npy')
    except FileNotFoundError:
        print("❌ 에러: x_train.npy, y_train.npy 파일이 없습니다.")
        return

    # 라벨 조정
    if Y_all.min() > 0:
        Y_all = Y_all - 1
        
    NUM_CLASSES = len(np.unique(Y_all))
    print(f"📊 데이터 정보: {len(X_all)}개 샘플, {NUM_CLASSES}개 클래스")

    # 마스크 생성
    Mask_all = (np.abs(X_all).sum(axis=2) > 0).astype(np.float32)

    # Split
    X_train, X_val, y_train, y_val, m_train, m_val = train_test_split(
        X_all, Y_all, Mask_all, test_size=0.2, random_state=42, stratify=Y_all
    )

    # 텐서 변환 (CPU에서는 .to(DEVICE)를 미리 하지 않고 DataLoader에서 함)
    train_dataset = TensorDataset(
        torch.FloatTensor(X_train), torch.LongTensor(y_train), torch.FloatTensor(m_train)
    )
    val_dataset = TensorDataset(
        torch.FloatTensor(X_val), torch.LongTensor(y_val), torch.FloatTensor(m_val)
    )

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

    # 모델 생성
    model = ChannelAgnosticNetwork(num_classes=NUM_CLASSES).to(DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

    # --- 학습 루프 ---
    print("\n🚀 학습 시작 (CPU 모드)...")
    start_time = time.time()
    
    for epoch in range(EPOCHS):
        model.train()
        train_loss = 0
        correct = 0
        total = 0
        
        for i, (x_batch, y_batch, mask_batch) in enumerate(train_loader):
            # 데이터를 CPU 장치로 보냄
            x_batch = x_batch.to(DEVICE)
            y_batch = y_batch.to(DEVICE)
            mask_batch = mask_batch.to(DEVICE)
            
            optimizer.zero_grad()
            outputs = model(x_batch, mask_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += y_batch.size(0)
            correct += (predicted == y_batch).sum().item()
            
            # CPU는 느리므로 진행상황을 자주 보여줌
            if (i+1) % 20 == 0:
                print(f"  [Epoch {epoch+1}] Batch {i+1}/{len(train_loader)} 진행 중...", end='\r')
            
        train_acc = 100 * correct / total
        
        # 검증
        model.eval()
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for x_batch, y_batch, mask_batch in val_loader:
                x_batch, y_batch, mask_batch = x_batch.to(DEVICE), y_batch.to(DEVICE), mask_batch.to(DEVICE)
                outputs = model(x_batch, mask_batch)
                _, predicted = torch.max(outputs.data, 1)
                val_total += y_batch.size(0)
                val_correct += (predicted == y_batch).sum().item()
        
        val_acc = 100 * val_correct / val_total
        epoch_time = time.time() - start_time
        
        print(f"\nTime: {epoch_time:.0f}s | Epoch [{epoch+1}/{EPOCHS}] "
              f"Loss: {train_loss/len(train_loader):.4f} | "
              f"Train Acc: {train_acc:.2f}% | "
              f"Val Acc: {val_acc:.2f}%")
        start_time = time.time() # 시간 초기화

    # 모델 저장
    torch.save(model.state_dict(), "gesture_model_cpu.pth")
    print("\n💾 학습 완료. 모델 저장됨: gesture_model_cpu.pth")

if __name__ == "__main__":
    train_model_cpu()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, Input, backend as K
from sklearn.model_selection import train_test_split
import os

# ====================================================
# 1. 모델 아키텍처 정의 (Keras Functional API)
# ==================
# ==================================
def build_cwe_ap_model(num_classes, d_model=128, input_channels=16, window_size=500):
    """
    Channel-wise Encoder + Attention Pooling 모델 (Keras 버전)
    """
    # -----------------------------------------------------
    # [입력] (Batch, 16, 500)
    # -----------------------------------------------------
    input_tensor = Input(shape=(input_channels, window_size), name='input_data')
    
    # Keras Conv1D는 (Time, Feature) 형태를 원하므로 
    # (16, 500) -> (16, 500, 1)로 차원 확장 (채널별로 1개의 신호가 있는 것으로 간주)
    x = layers.Reshape((input_channels, window_size, 1))(input_tensor)

    # -----------------------------------------------------
    # [Stage 1] Shared Encoder (Conv1D)
    # TimeDistributed를 사용하여 16개 채널에 '동일한' Encoder를 적용
    # -----------------------------------------------------
    # 인코더 모델 정의 (내부 구조)
    encoder_input = Input(shape=(window_size, 1))
    
    e = layers.Conv1D(32, kernel_size=15, padding='same', activation='relu')(encoder_input)
    e = layers.BatchNormalization()(e)
    e = layers.MaxPooling1D(2)(e)
    
    e = layers.Conv1D(64, kernel_size=5, padding='same', activation='relu')(e)
    e = layers.BatchNormalization()(e)
    e = layers.MaxPooling1D(2)(e)
    
    e = layers.Conv1D(d_model, kernel_size=3, padding='same', activation='relu')(e)
    e = layers.GlobalAveragePooling1D()(e) # 시간축 압축 -> (Batch, d_model)
    
    encoder_model = models.Model(encoder_input, e, name='shared_encoder')
    
    # 16개 채널 각각에 인코더 적용 -> 결과: (Batch, 16, d_model)
    features = layers.TimeDistributed(encoder_model)(x)

    # -----------------------------------------------------
    # [Stage 2] Attention Mechanism
    # -----------------------------------------------------
    # Attention Score 계산: (Batch, 16, 1)
    attn = layers.Dense(d_model // 2, activation='tanh')(features)
    attn_scores = layers.Dense(1)(attn) 
    
    # Softmax (채널 차원 axis=1에 대해 수행) -> (Batch, 16, 1)
    attn_weights = layers.Softmax(axis=1, name='attention_weights')(attn_scores)
    
    # Weighted Sum: Features * Weights -> (Batch, d_model)
    # (Batch, 16, d) * (Batch, 16, 1) -> (Batch, 16, d) -> Sum -> (Batch, d)
    weighted_features = layers.Multiply()([features, attn_weights])
    global_feature = layers.Lambda(lambda z: K.sum(z, axis=1))(weighted_features)

    # -----------------------------------------------------
    # [Classifier]
    # -----------------------------------------------------
    y = layers.Dense(64, activation='relu')(global_feature)
    y = layers.Dropout(0.3)(y)
    output = layers.Dense(num_classes, activation='softmax', name='output')(y)

    model = models.Model(inputs=input_tensor, outputs=output, name='CWE_AP_Model')
    return model

# ====================================================
# 2. 학습 실행 함수
# ====================================================
def train_with_tensorflow():
    print("🚀 TensorFlow로 학습을 준비합니다...")
    
    # CPU/GPU 확인
    print("사용 가능한 장치:", tf.config.list_physical_devices())

    # 1. 데이터 로드
    print("\n📂 1. 데이터 파일(.npy) 로드 중...")
    try:
        X_all = np.load('x_train.npy')
        Y_all = np.load('y_train.npy')
    except FileNotFoundError:
        print("❌ 에러: x_train.npy, y_train.npy 파일이 없습니다.")
        return

    # 라벨 조정 (0부터 시작하도록)
    if Y_all.min() > 0:
        Y_all = Y_all - 1
        
    num_classes = len(np.unique(Y_all))
    print(f"📊 데이터 정보: {len(X_all)} 샘플, {num_classes} 클래스")

    # Train / Validation 분할 (8:2)
    X_train, X_val, y_train, y_val = train_test_split(
        X_all, Y_all, test_size=0.2, random_state=42, stratify=Y_all
    )

    # 2. 모델 생성
    print("\n🏗️ 2. 모델 빌드 중 (Keras)...")
    model = build_cwe_ap_model(
        num_classes=num_classes,
        d_model=64,       # 경량화를 위해 줄임 (PC 사양 고려)
        input_channels=16,
        window_size=500
    )
    
    model.summary() # 모델 구조 출력

    # 3. 컴파일 및 학습
    print("\n🚀 3. 학습 시작...")
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy', # 정수형 라벨(0, 1, 2...)일 때 사용
        metrics=['accuracy']
    )

    # 학습 (Batch Size는 PC 사양에 맞춰 32 또는 64로 조절)
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=10,
        batch_size=32,
        verbose=1
    )

    # 4. 모델 저장
    model.save('cwe_ap_model_tf.h5')
    print("\n💾 모델 저장 완료: cwe_ap_model_tf.h5")

if __name__ == "__main__":
    # 텐서플로우가 설치되어 있는지 확인
    try:
        import tensorflow as tf
        train_with_tensorflow()
    except ImportError:
        print("❌ TensorFlow가 설치되지 않았습니다.")
        print("아래 명령어로 설치해주세요:")
        print("pip install tensorflow")
```
