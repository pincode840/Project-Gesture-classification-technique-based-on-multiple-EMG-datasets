```python
import pandas as pd
import numpy as np
from scipy import signal

def classify_and_upsample(df, target_col='Restimulus', original_hz=200, target_hz=1000):
    """
    데이터를 클래스별로 분류하고, 주파수를 업샘플링하는 함수
    
    Args:
        df: 원본 데이터프레임 (약 200만 행)
        target_col: 클래스가 들어있는 컬럼명 ('Restimulus')
        original_hz: 기존 샘플링 레이트 (200Hz)
        target_hz: 목표 샘플링 레이트 (1000Hz)
        
    Returns:
        pd.DataFrame: 업샘플링된 전체 데이터프레임
    """
    
    # 1. 업샘플링 비율 계산 (1000 / 200 = 5배)
    ratio = int(target_hz / original_hz)
    
    # 결과 데이터를 담을 리스트
    upsampled_chunks = []
    
    # 2. Restimulus(클래스) 별로 데이터 그룹화
    # groupby를 사용하여 클래스별로 데이터를 쪼갭니다.
    grouped = df.groupby(target_col)
    
    print(f"총 {len(grouped)}개의 클래스로 분류하여 업샘플링을 시작합니다...")
    
    for class_id, group_data in grouped:
        # 해당 클래스의 데이터 개수
        n_samples = len(group_data)
        
        # 목표 샘플 개수 계산
        new_n_samples = int(n_samples * (target_hz / original_hz))
        
        # 업샘플링 대상 컬럼 선택 (클래스 컬럼 제외한 EMG 데이터 컬럼들)
        feature_cols = group_data.columns.drop(target_col)
        features = group_data[feature_cols].values
        
        # 3. Scipy의 signal.resample을 사용하여 업샘플링 수행
        # (Fourier method: 신호의 주기성을 가정하고 주파수 영역에서 리샘플링)
        resampled_features = signal.resample(features, new_n_samples)
        
        # 4. 데이터프레임으로 재구성
        temp_df = pd.DataFrame(resampled_features, columns=feature_cols)
        
        # 클래스 라벨 다시 할당 (이 구간은 모두 동일한 클래스임)
        temp_df[target_col] = class_id
        
        upsampled_chunks.append(temp_df)
    
    # 5. 분리된 청크들을 하나로 병합
    final_df = pd.concat(upsampled_chunks, ignore_index=True)
    
    # 컬럼 순서 원상복구 (필요시)
    final_df = final_df[df.columns]
    
    return final_df

# ==========================================
# 사용 예시 (가상 데이터 생성 부분)
# ==========================================

ninapro_df['Restimulus'] = ninapro_df['Restimulus'].astype(int)

print(f"원본 데이터 크기: {ninapro_df.shape}")

# 2. 함수 실행
ninapro_df = classify_and_upsample(ninapro_df, target_col='Restimulus', original_hz=200, target_hz=1000)

print("작업 완료!")
print(f"업샘플링된 데이터 크기: {ninapro_df.shape}")
# 예상 크기: (2,023,639 * 5, 17) -> (10,118,195, 17)

# 결과 확인
print(ninapro_df.head())
```

```python
import pandas as pd
import numpy as np
from scipy import signal
import gc  # 가비지 컬렉터 (메모리 해제용)

def downsample_heavy_data(df, target_col='Restimulus', q=2):
    """
    메모리 오류 방지를 위해 groupby 대신 반복 필터링을 사용하는 다운샘플링 함수
    """
    
    # 1. 메모리 최적화: 전체 데이터를 float32로 변환 (이미 되어있다면 패스)
    # 메모리 사용량을 절반으로 줄임 (8 byte -> 4 byte)
    float_cols = df.select_dtypes(include=['float64']).columns
    if len(float_cols) > 0:
        print("메모리 절약을 위해 float64를 float32로 변환합니다...")
        df[float_cols] = df[float_cols].astype(np.float32)

    # 타겟 컬럼의 고유값(클래스) 추출
    unique_classes = df[target_col].unique()
    unique_classes.sort()
    
    downsampled_chunks = []
    
    print(f"총 {len(unique_classes)}개의 클래스에 대해 순차 처리를 시작합니다.")
    
    for cls in unique_classes:
        # 2. Groupby 대신 Boolean Indexing 사용
        # (Groupby 객체 생성 시 발생하는 거대한 메모리 오버헤드 방지)
        mask = (df[target_col] == cls)
        subset = df[mask]
        
        # 피처 추출
        feature_cols = subset.columns.drop(target_col)
        features = subset[feature_cols].values
        
        # 3. 다운샘플링 (decimate)
        # q=2 (2000Hz -> 1000Hz)
        try:
            resampled_data = signal.decimate(features, q=q, axis=0)
        except Exception as e:
            print(f"클래스 {cls} 처리 중 오류 발생: {e}")
            continue
            
        # 결과 데이터프레임 생성 (float32 유지)
        temp_df = pd.DataFrame(resampled_data, columns=feature_cols, dtype=np.float32)
        temp_df[target_col] = cls
        
        downsampled_chunks.append(temp_df)
        
        # 4. 메모리 강제 해제
        del subset
        del features
        del resampled_data
        del temp_df
        gc.collect() # 중요: 즉시 메모리 반환
        
        print(f"클래스 {cls} 완료", end='\r') # 진행상황 표시
    
    print("\n병합 중...")
    
    # 5. 최종 병합
    final_df = pd.concat(downsampled_chunks, ignore_index=True)
    
    # 컬럼 순서 원상복구 및 타입 정리
    final_df = final_df[df.columns]
    final_df[target_col] = final_df[target_col].astype(int)
    
    return final_df

# ==========================================
# 실행부 수정
# ==========================================

# [중요] 함수 호출 전에 원본 데이터 타입을 변환해도 좋습니다.
# nature_db의 용량을 미리 줄여놓고 함수에 넘기는 것이 가장 안전합니다.
print("메모리 최적화를 시작합니다 (컬럼별 순차 변환)...")

for col in nature_db.columns:
    # 해당 컬럼이 float64인 경우에만 변환
    if nature_db[col].dtype == 'float64':
        # 원본 데이터프레임의 컬럼을 직접 덮어씌움 (In-place)
        nature_db[col] = nature_db[col].astype(np.float32)
        print(f"컬럼 변환 완료: {col}", end='\r')

# 가비지 컬렉션으로 찌꺼기 메모리 정리
gc.collect()

print("\n모든 데이터가 float32로 변환되었습니다.")
print(f"현재 데이터 크기: {nature_db.shape}")

# 2. 이제 다운샘플링 함수 실행 (이전에 드린 최적화 함수 사용)
# 주의: 함수 호출 시 메모리 오류가 또 나면 커널(Kernel)을 재시작해야 할 수도 있습니다.
nature_db = downsample_heavy_data(nature_db, target_col='Restimulus', q=2)

print("작업 완료!")
# 2. 함수 실행
# 기존 변수(nature_db)를 덮어쓰지 않고 새로운 변수에 할당하여 메모리 피크 방지
nature_db = downsample_heavy_data(nature_db, target_col='Restimulus', q=2)

print("작업 완료!")
print(f"다운샘플링된 데이터 크기: {nature_db.shape}")
```

```python
import pandas as pd
import numpy as np
# import torch  <-- 에러의 원인이므로 주석 처리

def create_dataset_numpy_only(df, window_size=500, overlap_ratio=0.5, ignore_rest=True):
    """
    PyTorch 없이 Numpy만 사용하여 대용량 데이터를 전처리하는 함수
    """
    print("1. 데이터프레임을 Numpy로 변환 중...")
    # 메모리 절약을 위해 float32 사용
    # iloc[:, :16]은 emg1 ~ emg16 컬럼을 가정
    data_np = df.iloc[:, :16].values.astype(np.float32) 
    labels_np = df['Restimulus'].values.astype(np.int32)
    
    print("2. 동작 구간(Block) 분할 중...")
    # 라벨이 변하는 구간 탐지
    change_indices = np.where(np.diff(labels_np, prepend=labels_np[0]-1) != 0)[0]
    change_indices = np.append(change_indices, len(labels_np))
    
    X_list = []
    Y_list = []
    
    step_size = int(window_size * (1 - overlap_ratio))
    
    print(f"3. 슬라이딩 윈도우 수행 중 (Window: {window_size}, Step: {step_size})...")
    
    for i in range(len(change_indices) - 1):
        start_idx = change_indices[i]
        end_idx = change_indices[i+1]
        
        current_label = labels_np[start_idx]
        
        # 휴식(0) 제외
        if ignore_rest and current_label == 0:
            continue
            
        block_data = data_np[start_idx:end_idx]
        
        if len(block_data) < window_size:
            continue
            
        # 블록 내부 윈도우 슬라이싱
        for w_start in range(0, len(block_data) - window_size + 1, step_size):
            window = block_data[w_start : w_start + window_size]
            
            # (Time, Channel) -> (Channel, Time) 전치
            X_list.append(window.T) 
            Y_list.append(current_label)
            
    print("4. 결과 병합 중...")
    X_final = np.array(X_list, dtype=np.float32)
    Y_final = np.array(Y_list, dtype=np.int32)
    
    # 라벨 0부터 시작하도록 조정 (만약 1부터 시작한다면)
    if len(Y_final) > 0 and Y_final.min() > 0:
        Y_final = Y_final - 1
        
    print(f"완료! X shape: {X_final.shape}, Y shape: {Y_final.shape}")
    return X_final, Y_final

# ==========================================
# 실행 부분
# ==========================================

# 1. 함수 실행 (Torch 임포트 없이 진행)
x_train_np, y_train_np = create_dataset_numpy_only(
    learning_df, 
    window_size=500, 
    overlap_ratio=0.5
)

# 2. 결과 저장 (중요: PyTorch 고칠 때까지 데이터 보존)
np.save('x_train.npy', x_train_np)
np.save('y_train.npy', y_train_np)
print("데이터가 .npy 파일로 저장되었습니다.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, Input, backend as K
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import os
import gc # 가비지 컬렉션

# ====================================================
# 1. 모델 아키텍처 (동일)
# ====================================================
def build_cwe_ap_model(num_classes, d_model=128, input_channels=16, window_size=500):
    input_tensor = Input(shape=(input_channels, window_size), name='input_data')
    x = layers.Reshape((input_channels, window_size, 1))(input_tensor)

    encoder_input = Input(shape=(window_size, 1))
    e = layers.Conv1D(32, kernel_size=15, padding='same', activation='relu')(encoder_input)
    e = layers.BatchNormalization()(e)
    e = layers.MaxPooling1D(2)(e)
    e = layers.Conv1D(64, kernel_size=5, padding='same', activation='relu')(e)
    e = layers.BatchNormalization()(e)
    e = layers.MaxPooling1D(2)(e)
    e = layers.Conv1D(d_model, kernel_size=3, padding='same', activation='relu')(e)
    e = layers.GlobalAveragePooling1D()(e)
    
    encoder_model = models.Model(encoder_input, e, name='shared_encoder')
    features = layers.TimeDistributed(encoder_model)(x)

    attn = layers.Dense(d_model // 2, activation='tanh')(features)
    attn_scores = layers.Dense(1)(attn)
    attn_weights = layers.Softmax(axis=1, name='attention_weights')(attn_scores)
    
    weighted_features = layers.Multiply()([features, attn_weights])
    global_feature = layers.Lambda(lambda z: K.sum(z, axis=1))(weighted_features)

    y = layers.Dense(64, activation='relu')(global_feature)
    y = layers.Dropout(0.3)(y)
    output = layers.Dense(num_classes, activation='softmax', name='output')(y)

    return models.Model(inputs=input_tensor, outputs=output, name='CWE_AP_Model')

# ====================================================
# [핵심] 메모리 절약형 밸런싱 제너레이터
# 데이터를 미리 복사하지 않고, 학습할 때마다 각 클래스에서 균등하게 뽑아줍니다.
# ====================================================
class BalancedDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, x_data, y_data, batch_size=32):
        self.x = x_data
        self.y = y_data
        self.batch_size = batch_size
        self.classes = np.unique(y_data)
        self.num_classes = len(self.classes)
        
        # 각 클래스별 인덱스를 미리 저장 (메모리 사용량 미미함)
        self.class_indices = [np.where(y_data == c)[0] for c in self.classes]
        
        # 가장 많은 데이터가 있는 클래스 기준으로 epoch 길이 설정 (충분한 학습량 보장)
        max_samples = max([len(idx) for idx in self.class_indices])
        self.steps_per_epoch = max_samples // (batch_size // self.num_classes)

    def __len__(self):
        return self.steps_per_epoch

    def __getitem__(self, index):
        # 한 배치에 들어갈 각 클래스별 샘플 수 계산
        samples_per_class = self.batch_size // self.num_classes
        
        batch_x = []
        batch_y = []
        
        for i in range(self.num_classes):
            # 해당 클래스의 인덱스들 중 랜덤하게 선택 (중복 허용 = Oversampling 효과)
            indices = np.random.choice(self.class_indices[i], samples_per_class, replace=True)
            batch_x.append(self.x[indices])
            batch_y.append(self.y[indices])
        
        # 리스트를 numpy 배열로 변환
        batch_x = np.concatenate(batch_x, axis=0)
        batch_y = np.concatenate(batch_y, axis=0)
        
        # 배치 내부 셔플링 (중요: 클래스끼리 뭉쳐있지 않도록)
        idx = np.arange(len(batch_x))
        np.random.shuffle(idx)
        
        return batch_x[idx], batch_y[idx]

def train_with_tensorflow():
    print("🚀 TensorFlow 학습 준비 (Generator 방식)...")
    
    # 1. 데이터 로드 (메모리 절약을 위해 필요한 것만)
    try:
        X_all = np.load('x_train.npy').astype(np.float32) # float32로 로드
        Y_all = np.load('y_train.npy')
    except FileNotFoundError:
        print("❌ 데이터 파일을 찾을 수 없습니다.")
        return

    # 라벨 인코딩
    le = LabelEncoder()
    Y_all = le.fit_transform(Y_all)
    num_classes = len(np.unique(Y_all))
    
    # 2. 정규화 (Standard Scaling)
    print("⚖️ 데이터 정규화 중...")
    num_samples, channels, window = X_all.shape
    
    # (N, 16, 500) -> (N*500, 16) 형태로 변경하여 채널별 정규화
    X_permuted = np.transpose(X_all, (0, 2, 1)).reshape(-1, channels)
    scaler = StandardScaler()
    X_scaled_flat = scaler.fit_transform(X_permuted)
    
    # 원상 복구
    X_scaled = X_scaled_flat.reshape(num_samples, window, channels).transpose(0, 2, 1)
    
    # 메모리 정리 (중요)
    del X_all, X_permuted, X_scaled_flat
    gc.collect()
    
    print("✅ 정규화 완료.")

    # 3. Train / Validation 분할 (원본 데이터만 분할, 불균형 상태 유지)
    X_train, X_val, y_train, y_val = train_test_split(
        X_scaled, Y_all, test_size=0.2, random_state=42, stratify=Y_all
    )
    
    print(f"📊 학습 데이터셋 크기: {X_train.shape} (메모리 안전)")
    
    # ====================================================
    # [핵심] Generator 생성
    # ====================================================
    # 배치 사이즈: 19개 클래스가 최소 1~2개씩은 들어가야 하므로 64 이상 권장
    batch_size = 64 
    train_generator = BalancedDataGenerator(X_train, y_train, batch_size=batch_size)
    
    # 모델 생성
    model = build_cwe_ap_model(
        num_classes=num_classes,
        d_model=64,
        input_channels=channels,
        window_size=window
    )
    
    optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
    
    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    callbacks = [
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
        ModelCheckpoint('best_cwe_ap_model.keras', monitor='val_accuracy', save_best_only=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
    ]

    print("\n🚀 학습 시작 (Generator 사용)...")
    
    # fit 함수에 X, y 대신 generator를 넣습니다.
    history = model.fit(
        train_generator,
        validation_data=(X_val, y_val), # 검증 데이터는 그대로 사용
        epochs=100,
        callbacks=callbacks,
        verbose=1
    )

    print("\n✅ 학습이 완료되었습니다.")

if __name__ == "__main__":
    train_with_tensorflow()
```
